```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
# buradan sec istedigini: https://xeno-canto.org/
cikcik_url = "https://xeno-canto.org/437091/download"
!wget {cikcik_url} -O cikcik.mp3
```

```python
y, sr = librosa.load("cikcik.mp3")
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
librosa.display.waveshow(y, sr=sr, ax=ax[0])
ax[0].set(xlabel=None)
ax[0].set(ylabel="genlik")
librosa.display.specshow(D, y_axis='linear', x_axis='time', sr=sr, ax=ax[1])
ax[1].set(xlabel="zaman (s)")
plt.tight_layout()
plt.show()
```
